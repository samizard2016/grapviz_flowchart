```python
import graphviz
```

```python
f = graphviz.Digraph()
```

```python
steps = [f"step {i+1}" for i in range(5)]
```

```python
steps
```

```python
step_desc = [f'this is about step {i+1}' for i in range(5)]
```

```python
step_desc
```

```python
for idx,param in enumerate(zip(steps,step_desc)):
    if idx > 1:
        f.node(param[0],param[1],shape='box',style='filled',color='skyblue')
    else:
         f.node(param[0],param[1],shape='box',color='red')
    if idx > 0:
        f.edge(f"step {idx}",f"step {idx+1}",label=f'connecting\n the nodes',fontsize='6')
f.edge('step 1','step 3')
f.edge('step 4','step 1')
```

```python
ML = {
    "step 1":"Development Data",
    "step 2":"Initial set of Features",
    "step 3":"Is there any\n Missing Values",
    "step 4":"Data Imputations",
    "step 5":"EDA\n (Explortatory Data Analytics)",
    "step 6":"Determination of\n the Feature set",
    "step 7":"Feature Engineering\n - Data Transformations\n and Encoding",
    "step 8":"Selection of Features",
    "step 9":"Split Development data\n into Train, validation\n and test datasets",
    "step 10":"Development of the\n base model & the\n classification/Performance\n Report",
    "step 11":"Determination of the most\n effective Algorithm",
    "step 12":"Is the model\n good enough?",
    "step 13":"Fine tunig of\n the hyper parameters\n using brute-forces",
    "step 14":"Finalizing the model",
    "step 15":"Development of the\n restful API\n and the Docker Image",
    "step 16":"Delivery to ingestation"
}
```

```python
edgeLabs = {
                "step 1":"",
                "step 2":"scrutinizing\n the variables",
                "step 3":"YES",
                "step 4":"",
                "step 5":"data visualizations and\n inferential statistics",
                "step 6":"data visualizations",
                "step 7":"Lasso regularization",
                "step 8":"",
                "step 9":"with min of params",
                "step 10":"based on the\n performance values",
                "step 11":"Classification report\n /Performance Parameters",
                "step 12":"RandomizedSearchCV\n & GridSearchCV",
                "step 13":"",
                "step 14":"App & Container",
                "step 15":"Deployment & Ingestation"
            }
```

```python
mlf = graphviz.Digraph(filename='MLSteps.png')
for k,v in ML.items():
    step = int(k[-2:])
    if step > 5:
        if step==14:
            mlf.node(k,v,shape='diamond' if step==11 else 'box',style='filled',color='lightgreen')
        elif step==15:
            mlf.node(k,v,shape='diamond' if step==11 else 'box',style='filled',color='green')
        else:
            mlf.node(k,v,shape='diamond' if step==11 else 'box',style='filled',color='orange')        
    else:
        mlf.node(k,v,shape='diamond' if step==3 else 'box',style='filled', color='skyblue')
    if step < len(ML):
        mlf.edge(f"step {step}",f"step {step+1}",label=edgeLabs[f"step {step}"],fontsize='10',weight=3 if step=="10" else "1")
mlf.edge("step 3",'step 5',label="NO",fontsize="8")
mlf.edge("step 11",'step 5',label="NO",fontsize="8")
```
